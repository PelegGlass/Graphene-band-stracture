# Band stracture of graphene

## Lattice and vector basis

chosing the bravias lattive vectors:

$$\vec{a}_1 = a\left(\frac{\sqrt{3}}{2}, \frac{3}{2}\right), \quad
 \vec{a}_2 = a\left(-\frac{\sqrt{3}}{2}, \frac{3}{2}\right)$$

A sublattice atom at position $\vec{r}_n$ = $(0,1)$.

The nearest neigbhor vectors from the A atom are:

$$\vec{\delta_1} = a\left(0,1\right), \quad
\vec{\delta_2} = a\left(\frac{\sqrt{3}}{2}, -\frac{1}{2}\right), \quad
\vec{\delta_3} = a\left(\frac{-\sqrt{3}}{2}, -\frac{1}{2}\right)$$

So:

$$\vec{a_1} = \vec{\delta_1} - \vec{\delta_3}, \quad
\vec{a_2} = \vec{\delta_1} - \vec{\delta_2}$$


## Tight binding Hamiltonian


Only the nearest neigbors hopping are relevant.

Let $|\vec{R},A\rangle$ be the orbital on sublattice A in the unit cell at the Bravais vector $\vec{R}$, and similarly $|\vec{R},B\rangle$.


$$
  \hat{H}
  = -t \sum_{\mathbf{R}} \sum_{j=1}^{3}
      \Big(
         \ket{\mathbf{R},A}\bra{\mathbf{R} + \boldsymbol{\delta}_j,B}
       + \ket{\mathbf{R} + \boldsymbol{\delta}_j,B}\bra{\mathbf{R},A}
      \Big)
$$

- $t$ is the hopping amplitude (~2.7 eV for graphene)
- Assuming on site energies are 0.





## Bloch states 

Due to translation inveriance we go to **Bloch basis**:

$$
\begin{aligned}
  \ket{\mathbf{k},A}
  &= \frac{1}{\sqrt{N}} \sum_{\mathbf{R}}
      e^{i\mathbf{k}\cdot\mathbf{R}} \ket{\mathbf{R},A}, \\
  \ket{\mathbf{k},B}
  &= \frac{1}{\sqrt{N}} \sum_{\mathbf{R}}
      e^{i\mathbf{k}\cdot\mathbf{R}} \ket{\mathbf{R},B}
\end{aligned}
$$
where N is the number of unit cells.

## Hamiltonian in Bloch basis



In this basis, $\hat{H}$ becomes a $2\times2$ matrix for each $\vec{k}$:

$$H_{AA}(\mathbf{k}) = \bra{\mathbf{k},A}\hat{H}\ket{\mathbf{k},A}$$
$$H_{BB}(\mathbf{k}) = \bra{\mathbf{k},B}\hat{H}\ket{\mathbf{k},B}$$
$$H_{AB}(\mathbf{k}) = \bra{\mathbf{k},A}\hat{H}\ket{\mathbf{k},B}$$
$$H_{BA}(\mathbf{k}) = \bra{\mathbf{k},B}\hat{H}\ket{\mathbf{k},A}
$$
Because we onlu have hopping between A and B and we assumed no site energies:

$$
H_{AA}(\mathbf{k}) = H_{BB}(\mathbf{k}) = 0
$$

Now calculating $H_{AB}(\mathbf{k})$:
$$
\ket{\mathbf{k},B}
  = \frac{1}{\sqrt{N}} \sum_{\mathbf{R}}
      e^{i\mathbf{k}\cdot\mathbf{R'}} \ket{\mathbf{R'},B}
\\
      \bra{\mathbf{k},A}
  = \frac{1}{\sqrt{N}} \sum_{\mathbf{R}}
      e^{-i\mathbf{k}\cdot\mathbf{R}} \bra{\mathbf{R},A}
\\

$$

So:
$$
H_{AB}(\mathbf{k}) = \frac{1}{N}\sum_{\mathbf{R,R'}}
      e^{-i\mathbf{k}\cdot\mathbf{R}}e^{i\mathbf{k}\cdot\mathbf{R'}} \bra{\mathbf{R},A}\hat{H}\ket{\mathbf{R',B}}
$$
Reminder that $\hat{H}$ is of the form:
$$
-t\sum_{\mathbf{R,j}}\ket{R_0,A}\bra{R_0 + \delta_j, B} + h.c.
$$

Notice that all the nonzero matrix elemets are the neighbours:

$$
\bra{\mathbf{R},B}\hat{H}\ket{\mathbf{R'},A} = -t\sum_{j}\delta_{\mathbf{R},\mathbf{R'-\delta_j}}
$$

Plugging it in:

$$

H_{AB}(\mathbf{k})
= \frac{1}{N}\sum_{\mathbf{R},\mathbf{R}'}
e^{-i\mathbf{k}\cdot\mathbf{R}}\,
e^{i\mathbf{k}\cdot\mathbf{R}'}
\Bigl(-t \sum_{j} \delta_{\mathbf{R},\,\mathbf{R}'-\boldsymbol{\delta}_j}\Bigr) =

\frac{-t}{N}\sum_{\mathbf{R}',j}
e^{-i\mathbf{k}\cdot\mathbf{(R'-\delta_j)}}\,
e^{i\mathbf{k}\cdot\mathbf{R}'} = 

-t\sum_{j=1}^{3}
e^{i\mathbf{k}\cdot\mathbf{\delta_j}}

$$

Define:

$$
f(\mathbf{k}) = \sum_{j=1}^{3}e^{i\mathbf{k}\cdot\mathbf{\delta_j}}
$$

One could calculate $H_{BA}(\mathbf{k})$ and would get a similar answer but conjugate:

$$
H_{AB}(\mathbf{k}) = -tf(\mathbf{k}),\quad H_{Ba}(\mathbf{k}) = -tf^*(\mathbf{k})
$$

Therefore te Hamiltonian in the {${\ket{\mathbf{k},A}, \ket{\mathbf{k},B}}$} basis is :

$$
\hat{H}(\mathbf{k}) = 
\begin{pmatrix}
0 & -tf^*(\mathbf{k}) \\
-tf(\mathbf{k}) & 0
\end{pmatrix}
$$

Diagonalizing the $2\times2$ Hamiltonian will get you:

$$
E_\pm = \pm{t|f(\mathbf{k})|}
$$

## Computing $f(\mathbf{k})$ and $E_\pm{}(\mathbf{k})$

As we defined earlier:

$$\vec{\delta_1} = a\left(0,1\right), \quad
\vec{\delta_2} = a\left(\frac{\sqrt{3}}{2}, -\frac{1}{2}\right), \quad
\vec{\delta_3} = a\left(\frac{-\sqrt{3}}{2}, -\frac{1}{2}\right)$$

And let $\mathbf{k}=(k_x,k_y)$:
$$\mathbf{k}\cdot\mathbf{\delta_1}=ak_y$$
$$\mathbf{k}\cdot\mathbf{\delta_2}=a\left(\frac{\sqrt{3}}{2}k_x, \frac{1}{2}k_y\right)$$
$$\mathbf{k}\cdot\mathbf{\delta_2}=a\left(-\frac{\sqrt{3}}{2}k_x,- \frac{1}{2}k_y\right)$$

So:

$$
f(\mathbf{k}) = \sum_{j=1}^{3}e^{i\mathbf{k}\cdot\mathbf{\delta_j}} = 
e^{ik_ya} + e^{ia\left(\frac{\sqrt{3}}{2}k_x   -\frac{1}{2}k_y\right)} + e^{ia\left(-\frac{\sqrt{3}}{2}k_x - \frac{1}{2}k_y\right)}  =
e^{i k_y a}
+ e^{-i k_y a/2}
\left(
  e^{i a \frac{\sqrt{3}}{2} k_x}
  + e^{-i a \frac{\sqrt{3}}{2} k_x}
\right)

$$

Now using $e^{i\theta}+e^{-i\theta}=2\cos{\theta}$:

$$
f(\mathbf{k}) =
e^{i k_y a}
+ 2e^{-i k_y a/2}\cos{ \left(\frac{\sqrt{3}}{2} k_xa\right)}

$$

Now we can compute the eigen enrgies, Lets define for convenienve:

$$
\alpha = k_ya,\quad \beta = \cos{ \left(\frac{\sqrt{3}}{2} k_xa\right)}
$$
$$
f=e^{i\alpha}
+ 2e^{-i\alpha}\beta
\quad  
f^*=e^{-i\alpha}
+ 2e^{i\alpha}\beta

$$

$$
|f(\mathbf{k})| =ff^*  =  \left(e^{i\alpha}+ 2e^{-i\alpha}\beta\right)\left(e^{-i\alpha}+ 2e^{i\alpha}\beta\right)
=
1 + 2\beta{}\left(e^{i\frac{3\alpha}{2}} + e^{-i\frac{3\alpha}{2}}\right)
 + 4\beta{}^2
$$

And again using $e^{i\theta}+e^{-i\theta}=2\cos{\theta}$:

$$
|f(\mathbf{k})| = 1 + 4\beta{}\cos\left(\frac{3\alpha}{2}\right) + 4\beta{}^2
$$

Now substituting the values for $\alpha$ and $\beta$:

$$
|f(\mathbf{k})| = 1 + 4\cos{ \left(\frac{\sqrt{3}}{2} k_xa\right)}\cos\left(\frac{3k_ya}{2}\right) + 4\cos^2{ \left(\frac{\sqrt{3}}{2} k_xa\right)}
$$

So the energy bands are:

$$
\boxed{E_\pm{}(\mathbf{k}) = \pm{}t\sqrt{1 + 4\cos{ \left(\frac{\sqrt{3}}{2} k_xa\right)}\cos\left(\frac{3k_ya}{2}\right) + 4\cos^2{ \left(\frac{\sqrt{3}}{2} k_xa\right)}}}

$$



## Deriving the dirac points

The bands touch where $E(\mathbf{k}=0)$, so:
$$f(\mathbf{k}) = 0$$

Using our substitutions from  $\alpha = k_ya,\beta = \cos{ \left(\frac{\sqrt{3}}{2} k_xa\right)}$:

$$e^{i\alpha}+ 2e^{-i\alpha}\beta = 0 \Rightarrow e^{i\frac{3\alpha}{2}}+ 2\beta = 0 \Rightarrow e^{i\frac{3\alpha}{2}}= -2\beta $$

clearly, $|\beta|=\frac{1}{2}$, thus:

$$ \cos\left(\frac{\sqrt{3}}{2}k_xa \right) = \pm\frac{1}{2}$$

For $\beta = -\frac{1}{2}$:

$$ e^{\frac{3i\alpha}{2}} = 1$$

So:

$$  \frac{3\alpha}{2} = 2\pi{}n \Rightarrow  \alpha = \frac{4\pi{}n}{3}  \Rightarrow  k_y = \frac{4\pi{}n}{3a} $$

The smallest value $n=0 \Rightarrow k_y=0$ is required for the non equivalent dirac point ($n=\pm$ will get you the degenergy 6 dirac points in total).

$$ \cos\left(\frac{\sqrt{3}}{2}k_xa \right) = \frac{1}{2}  \Rightarrow \quad \frac{\sqrt{3}}{2}k_xa = \pm\frac{2\pi{}}{3} + 2\pi{}m$$

Taking $m=0$:

$$\mathbf{k} = \left(\frac{4\pi{}}{3\sqrt{3}a}, 0\right)  \quad \mathbf{k'} = \left(-\frac{4\pi{}}{3\sqrt{3}a}, 0\right)$$



## Plotting the energy bands


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  


t = 1.0
a = 1.0


nk = 200
kx = np.linspace(-np.pi, np.pi, nk)
ky = np.linspace(-np.pi, np.pi, nk)
KX, KY = np.meshgrid(kx, ky)

f = 1 + 4*np.cos((np.sqrt(3)/2) * a*KX)**2 + 4*np.cos((np.sqrt(3)/2) * a*KX)*np.cos((3/2) *a* KY)

E_plus  =  t * np.sqrt(f)
E_minus = -t * np.sqrt(f)

fig = plt.figure(figsize=(12, 12))
ax = plt.axes(projection='3d')

surf1 = ax.plot_surface(KX, KY, E_plus,  rstride=2, cstride=2, alpha=0.9)
surf2 = ax.plot_surface(KX, KY, E_minus, rstride=2, cstride=2, alpha=0.9)

k_x = (4*np.pi)/(3*np.sqrt(3)*a)

ax.scatter(k_x,0,0, c='black', marker='o',depthshade=False)
ax.scatter(-k_x,0,0, c='black', marker='o',depthshade=False)
ax.text(k_x,  0.1, 0.1,  r"$K$",  fontsize=10, color='k', zorder=11)
ax.text(-k_x, 0.1, 0.1,  r"$K'$", fontsize=10, color='k', zorder=11)
ax.set_xlabel(r"$k_x a$")
ax.set_ylabel(r"$k_y a$")
ax.set_zlabel(r"$E$")
ax.set_title("Graphene band structure")

plt.show()
```

## Band structure near Dirac points

Close to the dirac point $\mathbf{k = K+q}$, where $|q| \ll |K|$.

We saw that:

$$
f(\mathbf{k}) = \sum_{j=1}^{3}e^{i\mathbf{k}\cdot\mathbf{\delta_j}}  = 
\sum_{j=1}^{3}e^{i\mathbf{K+q}\cdot\mathbf{\delta_j}} =
\sum_{j=1}^{3}e^{i\mathbf{K}\cdot\mathbf{\delta_j}}e^{i\mathbf{q}\cdot\mathbf{\delta_j}}
$$

For small $\mathbf{q}$:

$$
e^{i\mathbf{q}\cdot\mathbf{\delta_j}} \approx 1+ i\mathbf{q\cdot\delta_j}
$$

Therefore:

$$
f(\mathbf{K+q}) \approx \sum_{j=1}^{3}e^{i\mathbf{K}\cdot\mathbf{\delta_j}}\left(1+ i\mathbf{q\cdot\delta_j}\right)= 
\underbrace{\sum_{j=1}^{3}e^{i\mathbf{K}\cdot\mathbf{\delta_j}} }_{=0}+ 
i\sum_{j=1}^{3}e^{i\mathbf{K}\cdot\mathbf{\delta_j}}\left(\mathbf{q\cdot\delta_j}\right)
$$

We saw earlier that where the bands touch $f\mathbf{k} = 0$, first sum is 0.
$$
f(\mathbf{K+q}) \approx i\sum_{j=1}^{3}e^{i\mathbf{K}\cdot\mathbf{\delta_j}}\left(\mathbf{q\cdot\delta_j}\right) = \frac{3a}{2}\left(-q_x+iq_y\right)
$$

Plugging it in the Hamiltonian from earlier:

$$
H(\mathbf{K+q}) \approx -t\frac{3a}{2}
\begin{pmatrix}
0 & -q_x+iq_y \\
-q_x-iq_y & 0
\end{pmatrix}
= \frac{3ta}{2}
\begin{pmatrix}
0 & q_x-iq_y \\
q_x+iq_y & 0
\end{pmatrix} = 
\frac{3ta}{2}\left(\sigma{}_xq_x+\sigma{}_yq_y\right)
$$

Where $\mathbf{\sigma}$ are the pauli matricies and $\mathbf{q}= (q_x,q_y,0)$:
$$
H(\mathbf{K+q}) = \frac{3ta}{2}(\mathbf{\sigma\cdot{}q})
$$

Now the eigen energies are imminennt:

$$
E(\mathbf{K+q}) = \pm\frac{3ta}{2}|\mathbf{q}|
$$

Where $|\mathbf{q}| = \sqrt{q_x^2 + q_y^2}$. 

## Plotting the band stracture near dirac points

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D   # needed for 3D plotting

t = 1.0
a = 1.0

nk = 200
qx = np.linspace(-np.pi, np.pi, nk)
qy = np.linspace(-np.pi, np.pi, nk)

QX, QY = np.meshgrid(qx, qy)

Q = np.sqrt(QX**2 + QY**2)

E_plus  =  (3 * t * a / 2.0) * Q
E_minus = -(3 * t * a / 2.0) * Q

fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111, projection='3d')


surf1 = ax.plot_surface(QX, QY, E_plus, alpha=0.9)
surf2 = ax.plot_surface(QX, QY, E_minus, alpha=0.9)



ax.scatter(0,0,0, c='black',s=30, marker='o',depthshade=False)
ax.text(0,  0.1, 0.1,  r"$K$",  fontsize=10, color='k', zorder=11)


ax.set_xlabel(r"$q_x$")
ax.set_ylabel(r"$q_y$")
ax.set_zlabel(r"$E$")

ax.set_title("Band stacture near dirac points")

plt.show()
```
